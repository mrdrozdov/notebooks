```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def compute_f1_score(xs, ys, label):
    xlabel, ylabel = xs == label, ys == label
    xtotal, ytotal = xlabel.sum(), ylabel.sum()
    matches = np.logical_and(xs == ys, ylabel)
    correct = matches.sum()
    prec = correct / xtotal
    rec = correct / ytotal
    return 2 * (prec * rec) / (prec + rec), prec, rec    
```

```python
np.random.seed(11)

n = 1000
points = np.random.randn(n, 2)
y = points[:, 0] > 0
x = np.random.randn(n) > 0

plt.figure()
plt.scatter(points[:, 0][y], points[:, 1][y], color='black')
plt.scatter(points[:, 0][np.logical_not(y)], points[:, 1][np.logical_not(y)], color='gray')
plt.scatter(points[:, 0][x], points[:, 1][x], marker='.', color='white')
plt.show()

print('f1 = {}, precision = {}, recall = {}'.format(*compute_f1_score(x, y, 1)))
```
